```python

%load_ext autoreload
%autoreload 2
import sys
# map folder using sys
sys.path.append('../')
from galgoz.backtesting import Backtest
from galgoz.indicators import WPR, SG, MFI, HMA, Hline, SuperTrend
from galgoz import DATA_FOLDER, Galgoz
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
bt = Backtest()
bt.run()
```

```python
gz = Galgoz()
indicators=gz.indicators(
    MFI(bt.data, window=11), 
    Hline(bt.data, yvalue=75), 
    Hline(bt.data, yvalue=25), 
    SuperTrend(bt.data, line_color="green"),
    )

fig = gz.plot_candles(bt.data, indicators=indicators, show=False, width=1400, height=800)
# Plot the BUY signals, labelled as 1 in the 'signals' column. Use the mid_l price to plot the signal and use a green triangle up marker.
fig.add_trace(go.Scatter(x=bt.data.time_str[bt.data['signals'] == 1], y=bt.data['mid_l'][bt.data['signals'] == 1], mode='markers', marker=dict(color='green', symbol='triangle-up', size=15), customdata=bt.data['signals'][bt.data['signals'] == 1], hovertemplate='Buy Signal', showlegend=False))
# Now SELL signals, labelled as -1 in the 'signals' column. Use the mid_l price to plot the signal and use a red triangle down marker.
fig.add_trace(go.Scatter(x=bt.data.time_str[bt.data['signals'] == -1], y=bt.data['mid_l'][bt.data['signals'] == -1], mode='markers', marker=dict(color='red', symbol='triangle-down', size=15), customdata=bt.data['signals'][bt.data['signals'] == -1], hovertemplate='Sell Signal', showlegend=False))
              
```
