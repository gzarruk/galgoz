```python

%load_ext autoreload
%autoreload 2
import sys
# map folder using sys
#sys.path.append('../')
import pandas as pd
import numpy as np
from galgoz import DATA_FOLDER, Galgoz
from galgoz.indicators import QQE, SuperTrend, Hline, MFI
from galgoz.utils import set_data_index_and_time_str
```

```python
filename = DATA_FOLDER / "GBP_JPY_H1.pkl"
data = pd.read_pickle(filename)
data = set_data_index_and_time_str(data)
```

```python
gz = Galgoz()
df_plot = data.iloc[-750:].copy()
indicators = gz.indicators(
    SuperTrend(df_plot, atr_period=21, multiplier=5.5), 
    Hline(df_plot, yvalue=50), 
    QQE(df_plot, length=8, smooth=5, factor=1.618)
    )
fig = gz.plot_candles(df_plot, instrument=gz.instrument, indicators=indicators, show=False)
fig.show()
```
