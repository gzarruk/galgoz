```python
%load_ext autoreload
%autoreload 2
import sys
# map folder using sys
#sys.path.append('../')
import pandas as pd
import numpy as np
from galgoz import DATA_FOLDER, Galgoz
from galgoz.indicators import QQE, SuperTrend, Hline, MFI, SG, HMA, WPR, RSI
from galgoz.utils import set_data_index_and_time_str
```

```python
filename = DATA_FOLDER / "GBP_JPY_H1.pkl"
data = pd.read_pickle(filename)
data = set_data_index_and_time_str(data)
```

```python
gz = Galgoz()
df_plot = data.iloc[-750:].copy()
indicators = gz.indicators(
    SuperTrend(df_plot, atr_period=21, multiplier=6), 
    SG(df_plot, line_color="maroon"),
    HMA(df_plot, line_color="purple"),
    Hline(df_plot, yvalue=0, line_color="black"), 
    WPR(df_plot, window=35, line_color="darkblue"),
    RSI(df_plot, window=14, line_color="darkred", row=3),
    QQE(df_plot, length=8, smooth=5, factor=1.618, row=4)
    )
fig = gz.plot_candles(df_plot, instrument=gz.instrument, indicators=indicators, show=False)
fig.show()
```
